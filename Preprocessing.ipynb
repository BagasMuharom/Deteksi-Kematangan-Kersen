```python
import cv2 as cv
import numpy as np
from Citra import Citra
from matplotlib import pyplot as plt
```

```python
plt.figure(figsize=(10,10))
image = cv.imread('data/tes.jpeg')
citra = Citra(image)
citra.toGray()
citra.toBinaryInv(127, 255)
contour = citra.getContour(9000, 1000000)
contour.croppedResize = (300, 300)
contour.findContour()

for i in contour.cropped:
    plt.imshow(cv.cvtColor(i, cv.COLOR_BGR2RGB))
# print(len(contour.cropped))
# plt.xticks([]), plt.yticks([])
plt.show()
```

```python
citra = Citra(cv.imread('data/tiga.jpg'))
citra.toGray()
citra.toBinaryInv(127, 255)
daftarGambar = citra.getContour(9000, 100000)
daftarGambar.croppedResize = (128, 128)
daftarGambar.findContour()

fig = plt.figure(figsize = (14, 14))
for counter, cropped in enumerate(daftarGambar.cropped):
    y = fig.add_subplot(6, 6, counter + 1)
    img = daftarGambar.cropped[counter]
    
    y.imshow(cv.cvtColor(img, cv.COLOR_BGR2RGB))
    plt.title(str(counter))
    y.axes.get_xaxis().set_visible(False)
    y.axes.get_yaxis().set_visible(False)
```
