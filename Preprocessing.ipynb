```python
import cv2 as cv
import numpy as np
from Citra import Citra
from matplotlib import pyplot as plt

citra = Citra(cv.imread('data/data mentah/baru/IMG_20181208_174620.jpg'))
citra.resize((1024, 768))
citra.toGray()
citra.toBinaryInv(150, 255)
cc = citra.getContour(4000, 100000)
cc.croppedResize = (128, 128)
cc.findContour()
fig = plt.figure(figsize = (14, 14))

# result = cv.cvtColor(cc.labeled, cv.COLOR_BGR2RGB)
# result = citra.processed
# plt.imshow(result, cmap='gray')
print('Jumlah buah : ' + str(len(cc.cropped)))

for counter in range(len(cc.cropped)):
    y = fig.add_subplot(10, 10, counter + 1)
    img = cc.cropped[counter]
    
    cv.imwrite('data/Dataset/v2/campur/' + str(counter + 45) + '.png', img)
    
    y.imshow(cv.cvtColor(img, cv.COLOR_BGR2RGB))
    plt.title(counter)
    y.axes.get_xaxis().set_visible(False)
    y.axes.get_yaxis().set_visible(False)
```

```python
import os

save = open("data/data learning/v2.txt", 'a')
path = 'data/Dataset/v2/4'
for file in os.listdir(path):
    if file == 'desktop.ini' or file == '.ipynb_checkpoints':
        continue
        
    citra = Citra(cv.imread(path + '/' + file))
    hist = citra.getHistogram()
    
    hist = ' '.join([str(x) for x in hist])
    save.write(hist + ' 0 0 0 1\n')
    
save.close()
```
