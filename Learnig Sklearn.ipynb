```python
from sklearn.neural_network import MLPClassifier
import numpy as np

dir = 'data/Learning.txt'
data = np.loadtxt(dir, delimiter=' ', dtype=int)

clf = MLPClassifier(solver='lbfgs', alpha=0, hidden_layer_sizes=(550, 350, 150), random_state=1, max_iter = 10000000, learning_rate_init=0.15)
clf.fit(data[:, :-1], data[:, -1])
clf.predict(data[:, :-1])
```
