```python
import cv2 as cv
from Citra import Citra

cap = cv.VideoCapture('data/data mentah/VID_20181126_174054.mp4')

while(True):
    ret, frame = cap.read()
    
    if ret == False:
        cap.set(2, 0)
        continue

    citra = Citra(frame)
    citra.resize((640, 320))
    citra.toGray()
    citra.toBinaryInv(127, 255)
    hasil = citra.getContour(9000, 100000)
    hasil.croppedResize = (128, 128)
    hasil.findContour()
    cv.imshow('frame', hasil.labeled)
    if cv.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv.destroyAllWindows()
```
