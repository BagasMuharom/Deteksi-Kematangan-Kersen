```python
import cv2 as cv
from Citra import Citra

cap = cv.VideoCapture('data/video.3gp')

while(True):
    ret, frame = cap.read()
    
    if ret == False:
        cap.set(2, 0)
        continue

    citra = Citra(frame)
    citra.resize((500, 300))
    citra.toGray()
    citra.toBinaryInv(127, 255)
    hasil = citra.getCountour()
    cv.imshow('frame', hasil['label'])
    if cv.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv.destroyAllWindows()
```
